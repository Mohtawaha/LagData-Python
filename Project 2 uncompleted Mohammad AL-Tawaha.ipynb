```python
sns.histplot(df[df['race'] == 'Caucasian']['age'], bins=20, kde=True, color='blue', label='Caucasian')
```

```python
plt.figure(figsize=(8, 6))
race_counts = df['race'].value_counts()
plt.bar(race_counts.index, race_counts.values)
plt.xlabel('Race')
plt.ylabel('Count')
plt.title('Race Distribution')
plt.show()
```
